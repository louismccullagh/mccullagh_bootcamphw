```python
import csv
import os
import pandas as pd
import numpy as np

school_data_csv = 'Resources/schools_complete.csv'
student_data_csv = 'Resources/students_complete.csv'

school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

### District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_num_schools = school_data["school_name"].count()
total_num_students = student_data["student_name"].count()
total_budget = complete_data["budget"].sum()
avg_math_score = complete_data["math_score"].mean()
avg_reading_score = complete_data["reading_score"].mean()

overall_passing_rate = ((avg_math_score + avg_reading_score)/2)

passing_math = complete_data.loc[(complete_data["math_score"] >= 70)]
num_passing_math = passing_math["math_score"].count()
percent_passing_math = (100*(num_passing_math/total_num_students))

passing_reading = complete_data.loc[(complete_data["reading_score"] >= 70)]
num_passing_reading = passing_reading["reading_score"].count()
percent_passing_reading = (100*(num_passing_reading/total_num_students))

Summarized_data = {"Total Number of Schools"                           : total_num_schools,
                   "Total Number of Students"                          : total_num_students,
                   "Total Budget"                                      : total_budget,
                   "Average Math Score"                                : avg_math_score,
                   "Average Reading Score"                             : avg_reading_score,
                   "Overall Passing Rate"                              : overall_passing_rate,
                   "Percentage of Students with Passing Math Score"    : percent_passing_math,
                   "Percentage of Students with Passing Reading Score" : percent_passing_reading,
                   }

Summarized_data_df = pd.DataFrame([Summarized_data])
Summarized_data_df = Summarized_data_df[["Total Number of Schools","Total Number of Students","Total Budget","Average Math Score","Average Reading Score","Overall Passing Rate","Percentage of Students with Passing Math Score","Percentage of Students with Passing Reading Score"]]
Summarized_data_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
School_Names = school_data["school_name"]
School_Type = school_data.set_index(["school_name"])["type"]
Num_Students_By_School = complete_data["school_name"].value_counts()
Budget_By_School = school_data.groupby(["school_name"]).sum()["budget"]
Budget_Per_Student = Budget_By_School/Num_Students_By_School
Avg_Math_Score_By_School = round(complete_data.groupby(["school_name"]).mean()["math_score"],2)
Avg_Reading_Score_By_School = round(complete_data.groupby(["school_name"]).mean()["reading_score"],2)

Overall_Passing_Rate_By_School = ((Avg_Math_Score_By_School + Avg_Reading_Score_By_School)/2)

passing_math_by_school = complete_data.loc[(complete_data["math_score"] >= 70)]
num_passing_math_by_school = passing_math_by_school.groupby(["school_name"]).count()["student_name"]
percent_passing_math_by_school = round((num_passing_math_by_school/Num_Students_By_School)*100,2)

passing_reading_by_school = complete_data.loc[(complete_data["reading_score"] >= 70)]
num_passing_reading_by_school = passing_reading_by_school.groupby(["school_name"]).count()["student_name"]
percent_passing_reading_by_school = round((num_passing_reading_by_school/Num_Students_By_School)*100,2)


Summarized_Data_By_School = pd.DataFrame({"School Type": School_Type,
                            "Number of Students by School": Num_Students_By_School,
                            "Budget by School": Budget_By_School,
                            "Budget Per Student": Budget_Per_Student,
                            "Average Math Score by School": Avg_Math_Score_By_School,
                            "Average Reading Score by School": Avg_Reading_Score_By_School,
                            "Overall Passing Rate by School": Overall_Passing_Rate_By_School,
                            "Percentage of Students with Passing Math Score by School": percent_passing_math_by_school,
                            "Percentage of Students with Passing Reading Score by School": percent_passing_reading_by_school
                            })
Summarized_Data_By_School
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
Summarized_Data_By_School["Scores by School Spending"] = pd.cut(Budget_Per_Student,spending_bins,labels = group_names)
```

```python
Math_Scores_By_Spending = round(Summarized_Data_By_School.groupby(["Scores by School Spending"]).mean()["Average Math Score by School"],2)
Reading_Scores_By_Spending = round(Summarized_Data_By_School.groupby(["Scores by School Spending"]).mean()["Average Reading Score by School"],2)
Percent_Passed_Math_by_Spending = round(Summarized_Data_By_School.groupby(["Scores by School Spending"]).mean()["Percentage of Students with Passing Math Score by School"],2)
Percent_Passed_Reading_by_Spending = round(Summarized_Data_By_School.groupby(["Scores by School Spending"]).mean()["Percentage of Students with Passing Reading Score by School"],2)
Overall_Passed_by_Spending = ((Percent_Passed_Math_by_Spending + Percent_Passed_Reading_by_Spending)/2)


Scores_by_Spending = pd.DataFrame({"Math Scores by Spending":Math_Scores_By_Spending,
                                     "Reading Scores by Spending":Reading_Scores_By_Spending,
                                     "Percent Passed Math by Spending":Percent_Passed_Math_by_Spending,
                                     "Percent Passed Reading by Spending":Percent_Passed_Reading_by_Spending,
                                     "Overall Passed by Spending":Overall_Passed_by_Spending})
Scores_by_Spending
```

## Scores by School Size
